```python
%matplotlib inline
%config InlineBackend.figure_formats = ["svg"]

import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import matplotlib.gridspec as gridspec

import seaborn as sns
sns.set_style('darkgrid')
np.random.seed(42)
```

```python
def univariate_normal(x, mean, variance):
    """ pd fof the univariate normal distribution """
    return ((1. / np.sqrt(2 * np.pi * variance)) * np.exp(-(x-mean)**2 / (2* variance)))
```

```python
x = np.linspace(-3, 5, num=100)
fig = plt.figure(figsize=(5, 3))
plt.plot(x, univariate_normal(x, mean=0, variance=1), label="$\mathcal{N}(0, 1)$")
plt.plot(x, univariate_normal(x, mean=2, variance=3), label="$\mathcal{N}(2, 3)$")
plt.plot(x, univariate_normal(x, mean=0, variance=0.2), label="$\mathcal{N}(0, 0.2)$")

# Why is value probability value > 1!?!?!?!
# plt.plot(x, univariate_normal(x, mean=0, variance=.1), label="$\mathcal{N}(0, .1)$")

plt.xlabel("$x$", fontsize=13)
plt.ylabel("Density: $p(x)$", fontsize=13)
plt.title("Univariate normal distribution")

plt.ylim([0, 1])
plt.xlim([-3, 5])
plt.legend(loc=1)

fig.subplots_adjust(bottom=0.15)
plt.show()
```

```python
univariate_normal(1, 0, .2)
```

```python
def multivariate_normal(x, d, mean, covariance):
    x_m = x - mean
    return (1/np.sqrt(np.power(2*np.pi, d)*np.linalg.det(covariance))) * np.exp(-(np.linalg.solve(covariance, x_m).T.dot(x_m))/2)
```

## Bi-variate normal distribution

```python
def generate_surface(mean, covariance, d):
    """
        Helper to generate density surface.
    """
    nb_of_x = 50 # grid size
    x1s = np.linspace(-5, 5, num=nb_of_x)
    x2s = np.linspace(-5, 5, num=nb_of_x)
        
    x1, x2 = np.meshgrid(x1s, x2s) # Generate grid
    pdf = np.zeros((nb_of_x, nb_of_x))
    
    for i in range(nb_of_x):
        for j in range(nb_of_x):
            pdf[i, j] = multivariate_normal(np.matrix([[x1[i,j]], [x2[i,j]]]), d, mean, covariance)
    
    return x1, x2, pdf
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 4))
d = 2

bivariate_mean = np.matrix(([0.], [0.])) # mean
bivariate_covariance = np.matrix([
    [1., 0.],
    [0., 1.]
])

x1, x2, p = generate_surface(bivariate_mean, bivariate_covariance, d)

# Plot bivariate distribution
con = ax1.contourf(x1, x2, p, 33, cmap=cm.YlGnBu)
ax1.set_xlabel("$x_1$", fontsize=13)
ax1.set_ylabel("$x_2$", fontsize=13)
ax1.axis([-2.5, 2.5 , -2.5, 2.5])
ax1.set_aspect("equal")
ax1.set_title("Independent variables", fontsize=12)


# Plot of correlated normals
bivariate_mean = np.matrix([[0.], [1.]]) # Mean
bivariate_covariance = np.matrix([
    [1., .8],
    [.8, 1.]
])
x1, x2, p = generate_surface(bivariate_mean, bivariate_covariance, d)
con = ax2.contourf(x1, x2, p, 33, cmap=cm.YlGnBu)
ax2.set_xlabel("$x_1$", fontsize=13)
ax2.set_ylabel("$x_2$", fontsize=13)
ax2.axis([-2.5, 2.5, -1.5, 3.5])
ax2.set_aspect("equal")
ax2.set_title("Correlated variables", fontsize=12)

# Add colorbar and title
fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([.85, .15, .02, .7])
cbar = fig.colorbar(con, cax=cbar_ax)
cbar.ax.set_ylabel("$p(x_1, x_2)$", fontsize=13)
plt.suptitle("Bivariate normal distributions", fontsize=13, y=.95)
plt.show()
```

## Sampling from a multivariate normal

```python
d = 2
mean = np.matrix([[0.], [1.]])
covariance = np.matrix([
    [1, .8],
    [.8, 1]
])

L = np.linalg.cholesky(covariance)

# Sample X from standard normal
n = 50
X = np.random.normal(size=(d, n))

# Apply affine transformation
Y = L.dot(X) + mean
```

```python
fig, ax = plt.subplots(figsize=(6, 4.5))

x1, x2, p = generate_surface(mean, covariance, d)
con = ax.contourf(x1, x2, p, 33, cmap=cm.YlGnBu)

# Plot samples
ax.plot(Y[0, :], Y[1, :], 'ro', alpha=.6, markeredgecolor='k', markeredgewidth=.5)
ax.set_xlabel("$y_1$", fontsize=13)
ax.set_ylabel("$y_2$", fontsize=13)
ax.set_aspect("equal")
ax.set_title("Samples from bivariate normal distribution")
cbar = plt.colorbar(con)
cbar.ax.set_ylabel("Density: $p(y_1, y_2)$", fontsize=13)
plt.show()
```

# Marginal & Conditional normal distributions

```python
d = 2 # Num of dimensions
mean = np.matrix([[0.], [1.]])
cov = np.matrix([
    [1, .8],
    [.8, 1]
])

# Get the mean values from the vector
mean_x = mean[0, 0]
mean_y = mean[1, 0]

# Get the blocks (single values in this case) from covariance matrix
A = cov[0, 0]
B = cov[1, 1]
C = cov[0, 1]
```

```python
fig = plt.figure(figsize=(7, 7))
gs = gridspec.GridSpec(2, 2, width_ratios=[2, 1], height_ratios=[2, 1])

plt.suptitle("Marginal distributions", y=.93)

ax1 = plt.subplot(gs[0])
x, y, p = generate_surface(mean, cov, d)
ax1.contourf(x, y, p, 33, cmap=cm.YlGnBu)
ax1.set_xlabel("$x$", fontsize=13)
ax1.set_ylabel("$y$", fontsize=13)
ax1.yaxis.set_label_position("right")
ax1.axis([-2.5, 2.5, -1.5, 3.5])

# Plot y
ax2 = plt.subplot(gs[1])
y = np.linspace(-5, 5, num=50)
py = univariate_normal(y, mean_y, A)

# Plot univariate distribution
ax2.plot(py, y, "b--", label=f'$p(y)$')
ax2.legend(loc=0)
ax2.set_xlabel("density", fontsize=13)
ax2.set_ylim(-1.5, 3.5)

# Plot x
ax3 = plt.subplot(gs[2])
x = np.linspace(-5, 5, num=50)
px = univariate_normal(x, mean_x, B)
ax3.plot(x, px, "r--", label="$p(x_1)$")
ax3.legend(loc=0)
ax3.set_ylabel("density", fontsize=13)
ax3.yaxis.set_label_position("right")
ax3.set_xlim(-2.5, 2.5)

# Clear axis 4 and plot colarbar in it


plt.show()
```
