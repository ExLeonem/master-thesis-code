# Max-Likelihood

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# Array indicating number of throws
n_dollar = np.asarray(np.linspace(0, 10, 11), dtype=int)
n_dollar
```

```python
# binom(k, n, p)
# k = number of success
# n = number of trials
# p = probability of single success
```

```python
# Probabilities of event occuring n-times
p_dollar_sign = binom.pmf(k=n_dollar, n=10, p=1/6)
p_dollar_sign
```

```python
plt.stem(n_dollar, p_dollar_sign)
plt.xlabel("Number of dollar sign")
plt.ylabel("Probability")
```

```python
# Probability of getting 2-dollar signs for n-throws
n_dollar = np.asarray(np.linspace(0, 6, 7), dtype=int)
p_dollar_signs = binom.pmf(k=2, n=10, p=n_dollar/6)
p_dollar_signs
```

```python
plt.stem(n_dollar, p_dollar_signs)
```

## Numerical Stability



```python
# Sample n-values from uniform distribution [0, 1]
vals_100 = np.random.uniform(0, 1, 100)
vals_1000 = np.random.uniform(0, 1, 1000)

# Multiple n-values
x_100 = np.product(vals_100)
x_1000 = np.product(vals_1000)

# Check values
print(x_100, x_1000)
```

```python
# Max-Min will stay at point-x on log transform
sample_count = 1000
vals = 1 + np.abs(np.sin(np.linspace(0, 3*np.pi, sample_count)) * np.sin(np.linspace(0, np.pi, sample_count)))

plt.plot(range(0, 1000), vals, 'b-')
plt.plot(range(0, 1000), np.log(vals), 'g--')
plt.xlabel("x")
plt.legend(("f(x)", "log(f(x))"), fontsize=10)
```

## Using logs for numerical stability

log(A * B) = log(A) + log(B)

```python
log_n_100 = np.sum(np.log(vals_100))
log_n_1000 = np.sum(np.log(vals_1000))

# Compare to above values. numerically more stable
print(log_n_100, log_n_1000)
```
