```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
BASE_PATH = os.path.join(os.getcwd(), "..", "..")
MODULES_PATH = os.path.join(BASE_PATH, "modules")
METRICS_PATH = os.path.join(BASE_PATH, "metrics")
STAT_PATH = os.path.join(BASE_PATH, "stats", "local")
```

```python
sys.path.append(MODULES_PATH)
sys.path.append(os.path.join(BASE_PATH, "tf_al"))
```

```python
%autoreload 2
from tf_al import ExperimentSuitMetrics
from utils import MetricsTransformer, Frame, ExperimentFrame, FrameScores, Stats
```

```python
dtypes = {
    "run": str,
    "iteration": int,
    "train_time": float,
    "eval_sparse_categorical_crossentropy": float,
    "eval_sparse_categorical_accuracy": float,
    "query_time": float,
    "labeled_pool_size": int,
    "unlabeled_pool_size": int,
    "method": str
}
```

```python
base_experiment_path = os.path.join(METRICS_PATH, "mnist_local")
metrics_handler = ExperimentSuitMetrics(base_experiment_path)
df_mc = MetricsTransformer.load_from_dir(metrics_handler, "mc_dropout", dtype=dtypes)
scores = FrameScores(accuracy_column="eval_sparse_categorical_accuracy")
mc_frame = ExperimentFrame(df_mc, scores=scores)
df_mc = mc_frame.get_frame()
```

```python
sel = df_mc["method"] == "BALD"
sns.lineplot(data=df_mc[sel], x="labeled_pool_size", y="eval_sparse_categorical_accuracy", hue="run")
```

```python
run_2 = np.logical_and(sel, df_mc["run"] == "2")
run_9 = np.logical_and(sel, df_mc["run"] == "9")
run_4 = np.logical_and(sel, df_mc["run"] == "4")
#run_6 = np.logical_and(sel, df_mc["run"] == "6")

sel_runs = np.logical_or(run_2, run_9)
sel_runs = np.logical_or(sel_runs, run_4)
#sel_runs = np.logical_or(sel_runs, run_6)

fig = sns.lineplot(data=df_mc[sel_runs], x="labeled_pool_size", y="eval_sparse_categorical_accuracy", hue="run")
fig.set(xlabel="Labeled pool size", ylabel="Test Accuracy")
fig.legend(title="Seeded-Run")
plt.savefig(os.path.join(STAT_PATH, "mc_acc_rise.png"))
```

```python
df_mp = MetricsTransformer.load_from_dir(metrics_handler, "moment_propagation", dtype=dtypes)
scores = FrameScores(accuracy_column="eval_sparse_categorical_accuracy")
mp_frame = ExperimentFrame(df_mp, scores=scores)
df_mp = mp_frame.get_frame()
```

```python
sel = df_mp["method"] == "BALD"
sns.lineplot(data=df_mp[sel], x="labeled_pool_size", y="eval_sparse_categorical_accuracy", hue="run")
```

```python
run_0 = np.logical_and(sel, df_mp["run"] == "0")
run_3 = np.logical_and(sel, df_mp["run"] == "1")
run_1 = np.logical_and(sel, df_mp["run"] == "9")

run_sel = np.logical_or(run_0, run_3)
run_sel = np.logical_or(run_sel, run_1)

fig = sns.lineplot(data=df_mp[run_sel], x="labeled_pool_size", y="eval_sparse_categorical_accuracy", hue="run")
fig.set(xlabel="Labeled Pool size", ylabel="Test Accuracy")
fig.legend(title="Seeded-Run")
plt.savefig(os.path.join(STAT_PATH, "mp_acc_rise.png"))
```
