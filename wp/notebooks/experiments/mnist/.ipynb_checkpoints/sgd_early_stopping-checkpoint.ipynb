```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BASE_PATH = os.path.join(os.getcwd(), "..", "..", "..")
MODULES_PATH = os.path.join(BASE_PATH, "modules")
METRICS_PATH = os.path.join(BASE_PATH, "metrics")
```

```python
sys.path.append(MODULES_PATH)
```

```python
from active_learning import ExperimentSuitMetrics
from utils import MetricsTransformer
```

```python
experiment_name = "3_08_2021_mnist_sgd_early_stopping"
path = os.path.join(METRICS_PATH, experiment_name)
```

```python
metrics_handler = ExperimentSuitMetrics(path)
```

```python
dtypes = {
    "run": int,
    "iteration": int,
    "train_time": float,
    "query_time": float,
    "eval_loss": float,
    "eval_accuracy": float,
    "labeled_pool_size": int
}
df_random = MetricsTransformer.load_from_dir(metrics_handler, "random", dtype=dtypes)
df_bald = MetricsTransformer.load_from_dir(metrics_handler, "bald", dtype=dtypes)
df_std_mean = MetricsTransformer.load_from_dir(metrics_handler, "std_mean", dtype=dtypes)
df_max_entropy = MetricsTransformer.load_from_dir(metrics_handler, "max_entropy", dtype=dtypes)
df_max_var_ratio = MetricsTransformer.load_from_dir(metrics_handler, "max_var_ratio", dtype=dtypes)
```

```python
df_bald.head()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 7))
plt.tight_layout(w_pad=5)
sns.lineplot(ax=ax[0], data=df_random, x="iteration", y="eval_accuracy")
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel("Eval. Accuracy")
ax[0].set_title("Method: Random")

sns.lineplot(ax=ax[1], data=df_bald, x="iteration", y="eval_accuracy")
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Eval. Accuracy")
ax[1].set_title("Method: BALD")

sns.lineplot(ax=ax[2], data=df_std_mean, x="iteration", y="eval_accuracy")
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Eval. Accuracy")
ax[2].set_title("Method: Std. Mean")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 7))

sns.lineplot(ax=ax[0], data=df_max_entropy, x="iteration", y="eval_accuracy")
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel("Eval. Accuracy")
ax[0].set_title("Method: Max Entropy")

sns.lineplot(ax=ax[1], data=df_max_var_ratio, x="iteration", y="eval_accuracy")
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Eval. Accuracy")
ax[1].set_title("Method: Max. Var. Ratio")
```

```python
dtypes = {
    "eval_loss": float,
    "eval_accuracy": float,
    "iteration": int,
    "labeled_pool_size": int,
    "query_time": float,
    "train_time": float
}
columns = list(dtypes.keys())
df_random_m = MetricsTransformer.mean(df_random, columns, dtype=dtypes)
df_bald_m = MetricsTransformer.mean(df_bald, columns, dtype=dtypes)
df_std_m = MetricsTransformer.mean(df_std_mean, columns, dtype=dtypes)
df_max_m = MetricsTransformer.mean(df_max_entropy, columns, dtype=dtypes)
df_var_m = MetricsTransformer.mean(df_max_var_ratio, columns, dtype=dtypes)

df_random_m.insert(0, "method", "Random")
df_bald_m.insert(0, "method", "BALD")
df_std_m.insert(0, "method", "Std. Mean")
df_max_m.insert(0, "method", "Max Entropy")
df_var_m.insert(0, "method", "Max Var Ratios")

dataframes = [df_random_m, df_bald_m, df_std_m, df_max_m, df_var_m]
```

```python
merged = pd.concat(dataframes, sort=True)
```

```python
merged.head()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 6))
sns.lineplot(ax=ax[0], data=merged, x="iteration", y="eval_accuracy", hue="method")
ax[0].set_xlabel("Iteration")
ax[0].set_ylabel("Eval. Accuracy")

sns.lineplot(ax=ax[1], data=merged, x="iteration", y="eval_loss", hue="method")
ax[1].set_xlabel("Iteration")
ax[1].set_ylabel("Eval. Loss")

sns.lineplot(ax=ax[2], data=merged, x="iteration", y="query_time", hue="method")
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Query Time")

sns.lineplot(ax=ax[3], data=merged, x="iteration", y="train_time", hue="method")
ax[3].set_xlabel("Iteration")
ax[3].set_ylabel("Train Time")
```
