```python
import os, sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
BASE_PATH = os.path.join(os.getcwd(), "..", "..", "..")
MODULES_PATH = os.path.join(BASE_PATH, "modules")
METRICS_PATH = os.path.join(BASE_PATH, "metrics")
```

```python
sys.path.append(MODULES_PATH)
```

```python
from active_learning import ExperimentSuitMetrics
from utils import MetricsTransformer
```

```python
path = os.path.join(METRICS_PATH, "6_08_2021_mnist_sgd_reset")
metrics_handler = ExperimentSuitMetrics(path)
```

```python
dtypes = {
    "iteration": int,
    "train_time": float,
    "eval_loss": float,
    "eval_accuracy": float,
    "query_time": float,
    "labeled_pool_size": int
}
df_random = MetricsTransformer.load_from_dir(metrics_handler, "random", dtype=dtypes)
df_bald = MetricsTransformer.load_from_dir(metrics_handler, "bald", dtype=dtypes)
df_ent = MetricsTransformer.load_from_dir(metrics_handler, "max_entropy", dtype=dtypes)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 10))

sns.lineplot(ax=ax[0], data=df_random, x="labeled_pool_size", y="eval_accuracy")
ax[0].set_xlabel("Pool Size")
ax[0].set_ylabel("Eval. Accuracy")
ax[0].set_title("Random")

sns.lineplot(ax=ax[1], data=df_ent, x="labeled_pool_size", y="eval_accuracy")
ax[1].set_xlabel("Pool Size")
ax[1].set_ylabel("Eval. Accuracy")
ax[1].set_title("Random")

sns.lineplot(ax=ax[2], data=df_bald, x="labeled_pool_size", y="eval_accuracy")
ax[2].set_xlabel("Pool Size")
ax[2].set_ylabel("Eval. Accuracy")
ax[2].set_title("Random")
```

```python
dtypes = {
    "iteration": int,
    "query_time": float,
    "train_time": float,
    "eval_accuracy": float,
    "eval_loss": float,
    "labeled_pool_size": int
}
columns = list(dtypes.keys())
df_random_mean = MetricsTransformer.mean(df_random, columns, dtype=dtypes)
df_ent_mean = MetricsTransformer.mean(df_ent, columns, dtype=dtypes)
df_bald_mean = MetricsTransformer.mean(df_bald, columns, dtype=dtypes)

df_random_mean.insert(0, "method", "Random")
df_ent_mean.insert(0, "method", "Max Entropy")
df_bald_mean.insert(0, "method", "BALD")

merged = pd.concat([df_random_mean, df_ent_mean, df_bald_mean])
```

```python
fig, ax = plt.subplots(1, 3, figsize=(20, 6))

sns.lineplot(ax=ax[0], data=merged, x="labeled_pool_size", y="eval_accuracy", hue="method")
ax[0].set_xlabel("Pool Size")
ax[0].set_ylabel("Eval. Accuracy")

sns.lineplot(ax=ax[1], data=merged, x="labeled_pool_size", y="eval_loss", hue="method")
ax[1].set_xlabel("Pool Size")
ax[1].set_ylabel("Eval. Loss")

sns.lineplot(ax=ax[2], data=merged, x="iteration", y="train_time", hue="method")
ax[2].set_xlabel("Iteration")
ax[2].set_ylabel("Time in seconds")

plt.show()
```
