```python
dtypes = {
    "run": int,
    "iteration": int,
    "train_time": float,
    "eval_sparse_categorical_crossentropy": float,
    "eval_sparse_categorical_accuracy": float,
    "query_time": float,
    "labeled_pool_size": int,
    "unlabeled_pool_size": int,
    "method": str
}
```

```python
base_experiment_path = os.path.join(METRICS_PATH, "mnist_local")
metrics_handler = ExperimentSuitMetrics(base_experiment_path)
df_mc = MetricsTransformer.load_from_dir(metrics_handler, "mc_dropout", dtype=dtypes)
scores = FrameScores(accuracy_column="eval_sparse_categorical_accuracy")
mc_frame = ExperimentFrame(df_mc, scores=scores)
df_mc = mc_frame.get_frame()
```

```python
def get_time(frame):
    selector = frame["model"] == "Moment Propagation"
    return max(frame[selector]["acc_times_min"])


def get_where(frame, model, acc, acc_key="eval_sparse_categorical_accuracy"):
    sel = np.logical_and(frame["model"] == model, frame[acc_key] >= acc)
    return np.array(frame[sel]["acc_times_min"])[0]
```

```python
def calc_total_time(frame):
    pool_size = frame["labeled_pool_size"]
    un_pool_size = frame["unlabeled_pool_size"]
    query_time = frame["query_time"]
    train_time = frame["train_time"]
    
    # Mean times in [Å¿]
    mean_qt = np.mean(query_time)
    
    return (un_pool_size*mean_qt)+train_time
```

```python
def sum_time(frame):
    
    total_time = calc_total_time(frame)
    acc_time = []
    for i in range(len(frame)):
        
        if i == 0:
            acc_time.append(total_time[i])
            continue
        
        summed_time = np.sum(total_time[:i+1])
        acc_time.append(summed_time)      
    
    return acc_time
```

```python
def create_new_frame(frame):
    methods = np.unique(frame["method"])
    sub_frames = []
    for method in methods:
        selector = frame["method"] == method
        acc_times = sum_time(frame[selector].copy())
        
        sub_frame = frame[selector].copy()
        sub_frame.insert(0, "acc_times", acc_times)
        sub_frames.append(sub_frame)
        
    return pd.concat(sub_frames)
```

```python
df_mc_mean = mc_frame.get_mean_frame()
df_mc_total_time = create_new_frame(df_mc_mean)
```

```python
%autoreload 2
from utils import ExperimentFrame
df_mp = MetricsTransformer.load_from_dir(metrics_handler, "moment_propagation", dtype=dtypes)
scores = FrameScores(accuracy_column="eval_sparse_categorical_accuracy")
mp_frame = ExperimentFrame(df_mp, scores=scores)
df_mp = mp_frame.get_frame()
```

```python
df_mp_total_time = create_new_frame(mp_frame.get_mean_frame())
```

```python
xlabel = "Total loop time in [min]"
ylabel = "Test Accuracy"
```

```python
composite = pd.concat([df_mp_total_time, df_mc_total_time], sort=True)
acc_times_min = composite["acc_times"]/60
composite.insert(0, "acc_times_min", acc_times_min)

sel = composite["method"] != "Random"
fig = sns.lineplot(data=composite[sel], x="acc_times_min", y="eval_sparse_categorical_accuracy", hue="method", style="model")

handle, labels = fig.get_legend_handles_labels()
labels[0] = labels[0].capitalize()
labels[-3] = labels[-3].capitalize()
fig.set(xlabel=xlabel, ylabel=ylabel)
fig.legend(handle, labels)

plt.savefig(os.path.join(STAT_PATH, "accumulated_times_mp_mc_all.png"), bbox_inches="tight")
```

```python
acc_key = "eval_sparse_categorical_accuracy"
sel = np.logical_and(composite["method"] == "BALD", composite[acc_key] > .8)
fig = sns.lineplot(data=composite[sel], x="acc_times_min", y=acc_key, hue="model")

nsp_mp = get_where(composite[sel], "Moment Propagation", .97)
nsp_mc = get_where(composite[sel], "MC Dropout", .97)
plt.axvline(x=nsp_mp, c="b", alpha=.4)
plt.axvline(x=nsp_mc, c="orange", alpha=.4)

handle, labels = fig.get_legend_handles_labels()
fig.set(xlabel=xlabel, ylabel=ylabel, ylim=(.9, 1))
fig.set_title("Method: BALD")
fig.legend(title="Model", loc="lower right")

plt.savefig(os.path.join(STAT_PATH, "acc_times_bald.png"), bbox_inches="tight")
```
